```python

# Programming Exercise1

# . Submit the GitHub link for your solution, which should contain the source code,

# datasets generated, and any other relevant material).
# Write a program to generate an Erdӧs-Renyi (E-R) network G(n, p), were for at least three input configurations, i.e., three set of
# values for n and p. For instance, one such configuration could be n = 7 ∗ 106
# and p = 10−2

# . For each configuration, calculate
# the following network properties and compare them with the theoretical prediction for an Erdӧs-Renyi network with the same
# configuration. Run each configuration at least 30 times and report the average estimates for each configuration.
# i. The average degree of the network.
# ii. Average clustering coefficient of the network.
# iii. The average path length.
# iv. Plot the degree distribution for the three chosen configurations. Do they center on the average degree? 
import math 
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def Average_degree(n,p):
    averageDegree = (n-1) * p
    return averageDegree
```

```python
def Average_clustering_coefficient(n,p):
    G = nx.erdos_renyi_graph(n,p)
    clustering = nx.algorithms.cluster.average_clustering(G)
    return clustering
```

```python

def Average_path_length(n,p):
    averageDegree = Average_degree(n,p)
    APL = ((math.log(n)) // (math.log(averageDegree)))
    return APL
```

```python
def Degree_distribution(n,p):
    G = nx.erdos_renyi_graph(n,p)
    degrees_dict = {}

    for n in G.nodes():
        degree = G.degree(n)
        if degree not in degrees_dict:
            degrees_dict[degree] = 0
        degrees_dict[degree] += 1
    items = sorted(degrees_dict.items())

    plt.plot([k for (k, v) in items], [v for (k, v) in items])
    
    plt.show()
    return items
```
